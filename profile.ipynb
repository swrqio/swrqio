```python
from PIL import Image, ImageDraw, ImageOps

# paths
banner_path = "assets/convert.png"
avatar_path = "assets/avatar_chibi.png"
out_path    = "assets/test.png"

# load
banner = Image.open(banner_path).convert("RGBA")
avatar = Image.open(avatar_path).convert("RGBA")

# target size & position to fit inside the circle ring
size = 240
x, y = 1150, 129

# resize avatar square, then make circular mask
avatar = avatar.resize((size, size), Image.LANCZOS)
mask = Image.new("L", (size, size), 0)
draw = ImageDraw.Draw(mask)
draw.ellipse((0, 0, size, size), fill=255)
avatar_circ = ImageOps.fit(avatar, (size, size), centering=(0.5, 0.5))
avatar_circ.putalpha(mask)

# paste into banner
banner.paste(avatar_circ, (x, y), avatar_circ)
banner.save(out_path)

print(f"Done -> {out_path}")
print(f"Avatar size: {size}px, Position: ({x}, {y})")
```
